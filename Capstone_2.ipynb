```python
import pandas as pd

shl_data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

shl_data
```

```python
shl_data.info()
```

```python
# Cleaning

# to see which are the categorical columns and see unique values
category = []
for col in shl_data.columns:
  if shl_data[col].dtype == 'O':
    category.append(col)
  else:
    pass

print(category)

for val in category:
  print('The unique value for ' + val + ' are:')
  print(shl_data[val].unique())
```

we can see that for the column 'BMI Category' "Normal" and "Normal Weight" are the same but done differently. Also, for 'Sleep Disorder' there are null values. We have to fix this!

```python
# replace 'Normal Weight' to 'Normal'

shl_data['BMI Category'] = shl_data['BMI Category'].replace(['Normal Weight'],['Normal'])

shl_data['BMI Category'].unique()
```

```python
# fill na values for 'Sleep Disorder'
shl_data['Sleep Disorder'] = shl_data['Sleep Disorder'].fillna('None')
shl_data['Sleep Disorder'].unique()
```

```python
# set 'Person ID' as the index

shl_data = shl_data.set_index('Person ID')

shl_data
```

```python
from bokeh.plotting import figure, curdoc, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, Select
from bokeh.layouts import column

output_notebook()
```

```python
source = ColumnDataSource(shl_data)

# Create scatter plot
plot = figure(width=600, height=400, title='Interactive Scatter Plot')
plot.circle(x='Age', y='Stress Level', source=source, size=10, color='navy', alpha=0.5)

show(plot)
```

```python
# Add dropdown menus
x_select = Select(title='X-axis:', value='Age', options=['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level'])
y_select = Select(title='Y-axis:', value='Stress Level', options=['Stress Level', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level'])

# Define callback function
def update_plot(attr, old, new):
    x = x_select.value
    y = y_select.value
    plot.xaxis.axis_label = x
    plot.yaxis.axis_label = y
    plot.circle(x=x, y=y, source=source, size=10, color='navy', alpha=0.5)

x_select.on_change('value', update_plot)
y_select.on_change('value', update_plot)

# Set up layout
layout = column(plot, x_select, y_select)

# Add layout to document
curdoc().add_root(layout)
```
