##Importing and Displaying Data

Data for this project is from Kaggle: [Electric Vehicle Population Data](https://www.kaggle.com/datasets/reenapinto/electric-vehicle-population-data/data)



```python
import pandas as pd
import numpy as np

ev_data = pd.read_csv("Electric_Vehicle_Population_Data.csv")

ev_data
```

Using the info method, we see that in this dataframe there is 150482 entries and 17 columns.

This is also tells us which columns has missing data, namely: County, City, Postal Code, Legislative District, Vehicle Location, Electric Utility, and 2020 Census Tract.


```python
ev_data.info()
```

For this project and the analysis I wanted to do, I will use the following columns and get rid of the rest.

**Make**: identifies the manufacturer or brand of the EV

**Model**: the model name of the EV

**Electric Vehicel Type**: categorizes the Ev into types such as battery electric vehicle (BEV) or plug-in hybrid electric vehicle (PHEV)

**Model Year**: indicates the year in which the EV was manufactured

**City**: the city where the EV is registered

**State**: the state where the EV is registered

**County**: the county where the EV is registered



##Preparing the Data

We know from getting the info above that the columns 'County' and 'City' are missing some values. To tackle this, I decided to drop the entire row where there are missing values for the two columns.

```python
ev_data = ev_data.dropna(subset=['County','City'])

ev_data.info()
```

Since there are no null values present in the dataframe, I will drop the columns I no longer need for my analysis.

```python
ev_data = ev_data[['City','State','County','Make','Model','Model Year','Electric Vehicle Type']]
print(ev_data.head())
```

Looking at the data preview above, I will keep the index as it is.



##Data Visualization

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()
```

Now, I want to segment electric vehicle market by make, model, and electric vehicle type. To do this, I will use the Group method.

```python
ev_make = ev_data.groupby('Make')
ev_model = ev_data.groupby('Model')
ev_type = ev_data.groupby('Electric Vehicle Type')
```

Visualize percentage of EV by make, model, and type.
To do this, I need to know the number of vehicle per make, model, and type.

```python
# by Make

import math

make_count = ev_make.size().tolist()
name = ev_data['Make'].unique()
make_name = np.sort(name)
source = {'Make': make_name, 'Count': make_count}

make_visual = figure(title = "Total Electric Vehicle by Make", x_range = make_name,
                     y_range = (0,80000), x_axis_label = 'Make', y_axis_label = 'Total EV',
                     height = 300, width=800)

make_visual.xaxis.major_label_orientation = math.pi / 4
make_visual.xgrid.grid_line_color = None

make_visual.vbar(x='Make', top='Count', width =0.7, source = source)
```

```python
show(make_visual)
```

This graph shows us that Chevyrolet, Nissan, and Tesla electric vehicles composed of most of the EVs that was registered. These are the only brands that has at least a 1000 units been registered.

```python
# by type

type_count = ev_type.size().tolist()
type_name = ev_data['Electric Vehicle Type'].unique()
type_source = {'Type': type_name, 'Count': type_count}

type_visual = figure(title = "Total Electric Vehicle by Type", x_range = type_name,
                     y_range = (0,120000), x_axis_label = 'Electric Vehicle Type',
                     y_axis_label = 'Total EV', height = 300, width=800)

type_visual.xgrid.grid_line_color = None

type_visual.vbar(x='Type', top='Count', width =0.7, source = type_source)
```

```python
show(type_visual)
```

This graph shows that there is more Battery Electric Vehicle that was registered in the state of Washington than Plug-in Hybrid Electric Vehicle by at least 50000 units.

```python
# by model

# since there are 37 EV makes in the dataframe, I will make a for loop that will see which brands
# have more than one type of electric vehicle to graph.

# Those that were not given a graph is because that make only has one type of electric vehicle;
# therefore I do not need to graph.

brands = [] # a list that contains the brands that has more than one type of EV

for brand in make_name:
  if ev_make.get_group(brand)['Model'].unique().size > 1:
    brands.append(brand)
  else:
    pass

from bokeh.layouts import gridplot

plots = []


for brand in brands:
  brand_data = ev_make.get_group(brand)

  model = brand_data['Model'].unique()

  model_fig = figure(x_range=model, title=brand, x_axis_label='Model',
                     y_axis_label ='Total EV')
  model_fig.xaxis.major_label_orientation = math.pi/4

  counts = list(brand_data['Model'].value_counts())

  data = {'Model': model, 'Count': counts}

  model_fig.vbar(x='Model', top='Count', width =0.8, source=data)

  plots.append(model_fig)

num_cols = 4
grid = gridplot(plots, ncols=num_cols, width=300, height=300)
show(grid)
```
