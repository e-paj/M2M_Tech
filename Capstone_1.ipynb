##Importing and Displaying Data

Data for this project is from Kaggle: [Electric Vehicle Population Data](https://www.kaggle.com/datasets/reenapinto/electric-vehicle-population-data/data)



```python
import pandas as pd

ev_data = pd.read_csv("Electric_Vehicle_Population_Data.csv")

ev_data
```

**Electric Vehicel Type**: categorizes the Ev into types such as battery electric vehicle (BEV) or plug-in hybrid electric vehicle (PHEV)

##Preparing the Data

We know from getting the info above that the columns 'County' and 'City' are missing some values. To tackle this, I decided to drop the entire row where there are missing values for the two columns.

```python
ev_data = ev_data.dropna(subset=['County','City'])

ev_data.info()
```

Since there are no null values present in the dataframe, I will drop the columns I no longer need for my analysis.

```python
ev_data = ev_data[['City','State','County','Make','Model','Model Year','Electric Vehicle Type']]
print(ev_data.head())
```

Looking at the data preview above, I will keep the index as it is. I could use VIN as my index, but as the goals does not need a specific vehicle, it is not necessary.



##Data Visualization

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import math

output_notebook()
```

Now, I want to segment electric vehicle market by make, model, and electric vehicle type. To do this, I will use the Group method.

###Visualize percentage of EV by make, model, and type.

To do this, I need to know the number of vehicle per make, model, and type.

```python
# by Make

ev_make = ev_data.groupby('Make').size().reset_index(name='Count')

# Calculate Percentage

make_total = ev_make['Count'].sum()
ev_make['Percentage'] = (ev_make['Count']/make_total)*100

source = ColumnDataSource(ev_make)

make_visual = figure(title = "Total Electric Vehicle by Make", x_range = ev_make['Make'],
                     x_axis_label = 'Make', y_axis_label = 'Total EV',
                     height = 300, width=800)

make_visual.xaxis.major_label_orientation = math.pi / 4
make_visual.xgrid.grid_line_color = None

make_visual.vbar(x='Make', top='Count', width =0.7, source = source)

# Add HoverTool
make_visual.add_tools(HoverTool(tooltips=[('Percentage','@Percentage%')]))
```

```python
show(make_visual)
```

This graph shows us that Chevyrolet, Nissan, and Tesla electric vehicles composed of most of the EVs that was registered. These are the only brands that has at least a 1000 units been registered.

```python
# by type

ev_type = ev_data.groupby('Electric Vehicle Type').size().reset_index(name='Count')

# Calculate percentage
total_type = ev_type['Count'].sum()
ev_type['Percentage'] = (ev_type['Count']/total_type)*100

type_source = ColumnDataSource(ev_type)

type_visual = figure(title = "Total Electric Vehicle by Type", x_range = ev_type['Electric Vehicle Type'],
                     x_axis_label = 'Electric Vehicle Type',
                     y_axis_label = 'Total EV', height = 300, width=800)

type_visual.xgrid.grid_line_color = None

type_visual.vbar(x='Electric Vehicle Type', top='Count', width =0.7, source = type_source)

type_visual.add_tools(HoverTool(tooltips=[('Percentage','@Percentage%')]))
```

```python
show(type_visual)
```

This graph shows that there is more Battery Electric Vehicle that was registered in the state of Washington than Plug-in Hybrid Electric Vehicle by at least 50000 units.

```python
# by model

# since there are 37 EV makes in the dataframe, I will make a for loop that will see which brands
# have more than one type of electric vehicle to graph.

# Those that were not given a graph is because that make only has one type of electric vehicle;
# therefore I do not need to graph.

brands = [] # a list that contains the brands that has more than one type of EV

make_group = ev_data.groupby('Make')
for brand in ev_make['Make']:
  if make_group.get_group(brand)['Model'].unique().size > 1:
    brands.append(brand)
  else:
    pass

from bokeh.layouts import gridplot

plots = []


for brand in brands:
  brand_data = make_group.get_group(brand)

  model = brand_data['Model'].unique()

  model_fig = figure(x_range=model, title=brand, x_axis_label='Model',
                     y_axis_label ='Total EV')
  model_fig.xaxis.major_label_orientation = math.pi/4

  counts = list(brand_data['Model'].value_counts())

  data = {'Model': model, 'Count': counts}

  model_fig.vbar(x='Model', top='Count', width =0.8, source=data)

  plots.append(model_fig)

num_cols = 4
grid = gridplot(plots, ncols=num_cols, width=300, height=300)
show(grid)
```

To see a bar graph of a specific Make, Model, and Electric Vehicle Type side by side:

```python
segmented_market = ev_data.groupby(['Make', 'Model', 'Electric Vehicle Type']).size().reset_index(name='Count')

segmented_market['Region']=segmented_market.apply(lambda row: f"{row['Make']} {row['Model']} ({row['Electric Vehicle Type']})", axis=1)

source = ColumnDataSource(segmented_market)

# Create a figure
segment_fig = figure(y_range=segmented_market['Region'].tolist(), height=1000, width = 1000,
                     title="Segmented Electric Vehicle Market", x_axis_label = "Number of Vehicles",
                     y_axis_label = "Make, Model, Electric Vehicle Type",
                     toolbar_location=None, tools="")


# Plot bars for each combination
segment_fig.hbar(y='Region',right='Count', height=0.5, source=source, color='skyblue')

segment_fig.ygrid.grid_line_color = None
```

```python
show(segment_fig)
```

###Analyze trends in electric vehicle sales by model year.

To do this:

**Group Data by Model Year**: Group the DataFrame by the 'Model Year' column.

**Count Vehicles per Model Year**: Count the number of vehicles for each model year.

**Plot the Trend**: Create a plot to visualize the trend in electric vehicle sales over different model years.

```python
# first change 'Model Year' from an integer to a string to make visualization easier

ev_data['Model Year'] = ev_data['Model Year'].apply(str)
```

```python
ev_data.info()
```

Group Data by Model Year:



```python
ev_year = ev_data.groupby('Model Year')
```

Count Vehicles per Model Year:


```python
sales_by_model_year = ev_year.size()
print(sales_by_model_year)
```

Plot the Trend:


```python
# Group the DataFrame by Model Year and count the number of vehicles for each year
sales_by_model_year = ev_data.groupby('Model Year').size().reset_index(name='Count')

# Calculate the percentage of sales for each model year
total_sales = sales_by_model_year['Count'].sum()
sales_by_model_year['Percentage'] = (sales_by_model_year['Count'] / total_sales) * 100

# Create a ColumnDataSource
source = ColumnDataSource(sales_by_model_year)

# Create a figure
year_visual = figure(x_range=sales_by_model_year['Model Year'].astype(str), height=500, title="Electric Vehicle Sales by Model Year",
           x_axis_label = "Model Year", y_axis_label = "Number of Vehicles",
           toolbar_location=None, tools="")

# Plot bars for each model year
year_visual.vbar(x='Model Year', top='Count', width=0.9, source=source, color='skyblue', alpha=0.8)

# Customize plot
year_visual.xaxis.major_label_orientation = 1.2
year_visual.y_range.start = 0
year_visual.ygrid.grid_line_color = None

# Add HoverTool with custom tooltip
year_visual.add_tools(HoverTool(tooltips=[('Percentage','@Percentage%')]))
```

```python
show(year_visual)
```

We can also use a line chart to see the trend better using the same info above.

```python
year_line = figure(title = "Electric Vehicle Sales by Model Year", x_range=sales_by_model_year['Model Year'].astype(str),
                   x_axis_label = 'Year',y_axis_label = 'Total EV', height = 300, width=800)

year_line.line(x='Model Year', y='Count', width =0.7, source = source, line_width=2)
year_line.ygrid.grid_line_color = None
```

```python
show(year_line)
```

###Utilize the columns such as County, City, and State to examine the distribution of electric vehicle models across different regions.

To do this:

**Group Data by Region**: Group the DataFrame by the desired geographical level (e.g., County, City, State).

**Count Vehicles per Region and Model**: Count the number of vehicles for each combination of region and electric vehicle model.

**Visualize the Distribution**: Create visualizations to illustrate the distribution of electric vehicle models across different regions.

Group Data by Region:

Group the DataFrame by County, City, and State, and count the number of vehicles for each combination

```python
vehicles_by_region = ev_data.groupby(['County','City','State','Model']).size().reset_index(name='Count')
```

Count Vehicles per Region and Model:

```python
# Create a list of unique regions by extracting unique combinations of 'County', 'City',
# and 'State' from the dataFrame created above:

regions = [tuple(region) for region in vehicles_by_region[['County', 'City', 'State']].values]
```

```python
# Have to do this way as the function (viridis) can only return colors up to the base palette's length (256)

from bokeh.palettes import viridis

# Determine the number of unique regions in dataset
num_regions = len(regions) # this returns the number of colors needed in our color palette

# Define the desired length of the palette
desired_palette_length = 256 # this is the number of colors to generate from the Viridis palette

# Calculate the number of times the palette needs to be repeated
repeat_factor = num_regions // desired_palette_length + 1
## Since the number of unique regions may be larger than the desired palette length,
## I calculated how many times I need to repeat the Viridis palette to cover all the regions.

# Generate the truncated palette and repeat it as needed
truncated_palette = viridis(desired_palette_length) * repeat_factor
## I generated a truncated version of the Viridis palette with the desired
## length. Then, I repeat this truncated palette as many times as necessary to cover all the regions.

# Use the truncated palette to assign colors to regions
palette = truncated_palette[:num_regions]
## Finally, I extract the first num_regions colors from the repeated palette.
## This ensures that I have enough colors to cover all the regions in the
##  dataset without encountering any errors.
```

Visualize the Distribution:

```python
# I have to do this with matplotlib because the notebook had a hard time displaying the Bokeh plot for this figure

import matplotlib.pyplot as plt

# Visualize the distribution of electric vehicle models across different regions
plt.figure(figsize=(50, 25))

for region, group in vehicles_by_region.groupby(['County', 'City', 'State']):
    plt.bar(group['Model'], group['Count'], label=f"{region[0]}, {region[1]}, {region[2]}")

plt.xlabel('Electric Vehicle Model')
plt.ylabel('Number of Vehicles')
plt.title('Distribution of Electric Vehicle Models Across Different Regions')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```
